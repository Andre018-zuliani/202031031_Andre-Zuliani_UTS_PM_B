### Mengubah object menjadi bentuk Numerik
#### Disini saya menginput beberapa type data object yang sudah kita preprocesseing menggunakan data.info()

```python
from sklearn.preprocessing import LabelEncoder
for col in ['Sex','ChestPainType','FastingBS','RestingECG','ExerciseAngina', 'ST_Slope', 'HeartDisease']:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    data
    print (data['Sex'],data['ChestPainType'],data['FastingBS'],data['RestingECG'],data['ExerciseAngina'],data['ST_Slope'],data['HeartDisease'])
```

### Menentukan data x dan y

```python
x = data.drop('HeartDisease', axis = 1) #Disini saya menginput kolom HeartDisease pada variable x dan y
y = data['HeartDisease']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state =0)
```

```python
from sklearn.neighbors import KNeighborsClassifier #disini saya mengimport KNegihborsClassifier untuk pemodelan KNN

knn = KNeighborsClassifier(n_neighbors =1)
```

### Pemodelan KNN

```python
knn.fit(x_train, y_train) #Mencari nilai n_Neighbors
```

```python
print("Akurasi score sebesar: {}".format(knn.score(x_test,y_test)*100)) # disini saya menampilkan hasil akurasi score yang sebesar 67%
```
