```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
import math

dataset = pd.read_csv('R04_corruptions.csv')
dataset.head()
```

```python
dataset.describe() #mendeskripsi dataset
```

#### disini saya memberikan atribut pada kolom Corruption Percetion ke variable X dan pada kolom Government Transparency ke variable Y.
#### lalu saya ingin menampilkan 2 variable yang sudah saya definisikan dengan plt.show()

```python
dataset.plot(x='Corruption Perception (X)', y='Government Transparency (Y)', style='o')
plt.title('Corruption Perception vs Government Transparency')
plt.xlabel('Corruption Perception')
plt.ylabel('Government Transparency')
plt.show()
```

### Korelasi

```python
dataset.corr() #Berikut Korelasi antar kedua data
```

```python
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

#### Disini saya membuat variable baru yaitu x_train dan y_train, lalu saya membuat variable regressor 

```python
print(regressor.intercept_)
```

```python
print(regressor.coef_)
```

### perhitungan manual dengan rumus : y = intercept + coef 

```python
99.94197593599009 + -10.82303393 
```

```python
y_pred = regressor.predict(x_test)
```

### Dataframe

```python
dataframe = pd.DataFrame({'Data Sebenarnya' : y_test, 'Data Prediksi' : y_pred}) #Membuat dataframe yang dimana saya membagi antara data sebenarnya dan data prediksi
dataframe
```

### Visualisasi hasil Test Set

```python
plt.scatter(x_test, y_test, color = "Green")
plt.plot(x_train, regressor.predict(x_train), color = 'yellow')
plt.title('Corruption Perception vs Government Transparency (Test Set)')
plt.xlabel('Corruption Perception')
plt.ylabel('Government Transparency')
plt.show()
```

```python
plt.scatter(x_train, y_train, color = "Green")
plt.plot(x_train, regressor.predict(x_train), color = 'yellow')
plt.title('Corruption Perception vs Government Transparency (Train Set)')
plt.xlabel('Corruption Perception')
plt.ylabel('Government Transparency')
plt.show()
```
